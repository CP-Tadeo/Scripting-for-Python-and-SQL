```python
import mysql.connector

```

```python
# Replace these with your actual database credentials
host = 'localhost'
user = 'root'
password = 'root'
database = 'misc'
```

```python
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database
)
```

```python
import pandas as pd

csv_file = 'National-dex.csv'


df = pd.read_csv(csv_file, index_col=0)
headers_list = df.columns.tolist()
```

```python
cursor = connection.cursor()
```

```python
create_table_query = '''
CREATE TABLE pokemon(
    dex_num INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(255),
    Type1 VARCHAR(255),
    Type2 VARCHAR(255),
    Ability1 VARCHAR(255),
    Ability2 VARCHAR(255),
    Ability3 VARCHAR(255),
    HealthPoints INT,
    Attack INT,
    Defense INT,
    SpecialAttack INT,
    SpecialDefense INT,
    Speed INT
)
'''
```

```python
cursor = connection.cursor()
cursor.execute(create_table_query)
connection.commit()
```

```python
for _, row in df.iterrows():
    # Exclude 'dex_num' from the columns and values
    columns = ', '.join(df.columns)
    values = ', '.join(['%s'] * len(df.columns))
    insert_query = f"INSERT INTO pokemon ({columns}) VALUES ({values})"
    cursor.execute(insert_query, tuple(row))
#the commit is always important. it is what finalizes the code in the database
connection.commit()
```

```python
# Execute the SELECT query and fetch the results
select_query = 'SELECT * FROM pokemon;'
cursor.execute(select_query)

#  Fetch and print the results
results = cursor.fetchall()
for row in results:
    print(row)
```
